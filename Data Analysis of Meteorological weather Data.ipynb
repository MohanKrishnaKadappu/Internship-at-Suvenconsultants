```python
# We have too convert the Formatted Date datatype from object to datetype, here we are also telling UTC=True to Pandas that your dates and times should not be naive, but UTC.
my_data['Formatted Date'] = pd.to_datetime(my_data['Formatted Date'],utc=True)
```

```python
my_data = my_data.set_index('Formatted Date')
my_data.head()
```

Here the given data[Apparent temperature and Humidity] is in hour period, hence we have to resample the given data to monthly for a superior analysis. 

```python
#resample() is a method of pandas dataframes that can be used to summarize data by date or time. The .mean() method will add up all values for each resampling period (ex for each month) to provide a mean[avg] output value for that period.
my_data_monthly_sample = my_data[['Apparent Temperature (C)','Humidity']].resample('M').mean()
my_data_monthly_sample.head()
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(18, 8), dpi=80, facecolor='lavender', edgecolor='tan')
sns.lineplot(x="Apparent Temperature (C)",y="Humidity",data=my_data_monthly_sample)
plt.title("Relation of Apparent Temperature (C) and Humidity")
```

# II ] Plot of Apparent Temperature and Humidity with year

```python
plt.figure(figsize=(18,8))
plt.plot(my_data_monthly_sample['Humidity'],label='Humidity with year',color='navy')
plt.plot(my_data_monthly_sample['Apparent Temperature (C)'],label='Apparent Temperature (C)',color='magenta')
plt.xlabel("Year",fontsize="16")
plt.title("Plot of Apparent Temperature and Humidity with year")
plt.legend(loc=5,fontsize=11)
```
